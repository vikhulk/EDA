```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 15,8
p = sns.countplot(DF['TypeName'], palette = 'pastel')
plt.title('Laptop Brands', fontsize = 20)
plt.xlabel('Typename')
plt.ylabel('Count')
plt.show()
```
